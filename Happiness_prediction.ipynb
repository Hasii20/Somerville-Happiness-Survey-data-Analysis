### Creating Datasets Based on Survey Year 

```python
unique_years = df['Year'].unique()

df_dict = {}

for year in unique_years:
    df_year = df[df['Year'] == year]
    # Drops columns which are completely empty.
    df_year_PP = df_year.dropna(axis=1, how='all')
    # Drops 'Year' column.
    df_year_PP = df_year_PP.drop(columns=['Year'])
    df_dict[f'df_{year}'] = df_year_PP
```

### Preprocessing df_2023

### Dropping variables with missing value percentage > 50% 

```python
missing_per = df_2023.isnull().mean() * 100
missing_per
remove_variables = missing_per[missing_per >50].index
remove_variables
```

```python
df_2023.drop(remove_variables, axis =1, inplace = True)
```

### Dropping variables that were used to derive other variables

```python
# Since Income.Per.Number.In.Household is obtained by dividing Household.Income.mid by Household.people.num
```

```python
df_2023.drop(columns = ['Household.Income.mid','Household.Size.num'], axis =1, inplace = True)
```

```python
# Since Rent.as.Percent.of.Income is obtained by dividing Rent.Mortgage.mid (multiplied by twelve for an annual value) by Household.Income.num
```

```python
df_2023.drop(columns = ['Rent.Mortgage.mid'], axis =1, inplace = True)
```

```python
# Since Rent.Mortgage.Per.Bedroom is obtained by dividing Rent.Mortgage.mid by Bedrooms.num
```

```python
df_2023.drop(columns = ['Bedrooms.num'], axis =1, inplace = True)
```

### Dropping missing values in target variable 

```python
df_2023.dropna(subset=['Happiness.5pt.num'], inplace=True)
```

### Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x = df_2023.drop('Happiness.5pt.num', axis=1)
y = df_2023['Happiness.5pt.num']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
y.info
```

### Checking for distributions of continuous variables

```python
import matplotlib.pyplot as plt

continuous_columns = ['Income.Per.Number.In.Household', 'Rent.as.Percent.of.Income', 'Rent.Mortgage.Per.Bedroom']

num_plots = len(continuous_columns)
fig, axs = plt.subplots(num_plots, 1, figsize=(10, 6*num_plots))

for i, column in enumerate(continuous_columns):
    axs[i].hist(x_train[column], bins=20, color='skyblue', edgecolor='black')
    axs[i].set_title(f'Histogram of {column}')
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency')
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
for col in continuous_columns:
    skewness=x_train[col].skew()
    print(f'{col} has skewness : {skewness}')
```

### Imputing continuous variables 

```python
medians = x_train[continuous_columns].median()
median_values = {column: medians[column] for column in continuous_columns}
x_train.fillna(median_values, inplace=True)
x_test.fillna(median_values, inplace=True)
```

### Imputing categorical variables

```python
for column in x_train.columns:
    if column not in continuous_columns:
        if x_train[column].isnull().any():
            mode_value = x_train[column].mode()[0]
            x_train[column].fillna(mode_value, inplace=True)
            x_test[column].fillna(mode_value, inplace=True)
```

```python
print(x_train['Gender'].value_counts())
print(x_train['Race.Ethnicity'].value_counts())
print(x_train['Housing.Status'].value_counts())
print(x_train['Days.In.Person.Commute'].value_counts())
print(x_train['Language.Spoken'].value_counts())
```

```python
gender_mapping = {
    'Prefer not to answer': 'Other',
    'Nonbinary': 'Other',
    'Multiple people': 'Other'
}

x_train['Gender'] = x_train['Gender'].replace(gender_mapping)
x_test['Gender'] = x_test['Gender'].replace(gender_mapping)

race_mapping = {
    'Prefer not to answer': 'Other',
    'Two or more races': 'Other',
    'Another race': 'Other'
}

x_train['Race.Ethnicity'] = x_train['Race.Ethnicity'].replace(race_mapping)
x_test['Race.Ethnicity'] = x_test['Race.Ethnicity'].replace(race_mapping)

house_mapping = {
    'Prefer not to answer': 'Other',
}

x_train['Housing.Status'] = x_train['Housing.Status'].replace(house_mapping)
x_test['Housing.Status'] = x_test['Housing.Status'].replace(house_mapping)
```
